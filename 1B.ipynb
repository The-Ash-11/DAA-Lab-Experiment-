```python
import random

def interpolation_search(roll_numbers, target):
    """
    Interpolation Search adapted for student roll numbers.
    Returns: (index_found, number_of_probes)
    """
    low = 0
    high = len(roll_numbers) - 1
    probes = 0

    while low <= high and roll_numbers[low] <= target <= roll_numbers[high]:
        probes += 1

        if low == high:
            if roll_numbers[low] == target:
                return low, probes
            return -1, probes

        # Interpolation formula to guess the position
        pos = low + int(((target - roll_numbers[low]) * (high - low)) /
                        (roll_numbers[high] - roll_numbers[low]))

        if roll_numbers[pos] == target:
            return pos, probes
        elif roll_numbers[pos] < target:
            low = pos + 1
        else:
            high = pos - 1

    return -1, probes


def binary_search(roll_numbers, target):
    """
    Binary Search for comparison.
    Returns: (index_found, number_of_probes)
    """
    low = 0
    high = len(roll_numbers) - 1
    probes = 0

    while low <= high:
        probes += 1
        mid = (low + high) // 2

        if roll_numbers[mid] == target:
            return mid, probes
        elif roll_numbers[mid] < target:
            low = mid + 1
        else:
            high = mid - 1

    return -1, probes


def compare_probe_performance():
    """
    Simulates a database of student roll numbers and compares the
    number of probes required by both search algorithms.
    """
    # Simulate a realistic scenario: 8,500 enrolled students out of a 10,000 capacity
    # This creates a mostly uniform but slightly gapped array of roll numbers
    MAX_ROLL = 10000
    ENROLLED_STUDENTS = 8500

    roll_numbers = sorted(random.sample(range(1, MAX_ROLL + 1), ENROLLED_STUDENTS))

    # Select a few random test cases from the array
    targets_to_test = [
        roll_numbers[random.randint(0, len(roll_numbers) - 1)] for _ in range(5)
    ]

    print("--- Roll Number Search Probe Comparison ---")
    print(f"Total Enrolled Students: {ENROLLED_STUDENTS}")
    print(f"Roll Number Range: 1 to {MAX_ROLL}\n")

    print(f"{'Target Roll No.':>16} | {'IS Probes':>10} | {'BS Probes':>10} | {'Winner':>12}")
    print("-" * 57)

    for target in targets_to_test:
        idx_is, probes_is = interpolation_search(roll_numbers, target)
        idx_bs, probes_bs = binary_search(roll_numbers, target)

        # Determine which algorithm was more efficient
        if probes_is < probes_bs:
            winner = "Interpolation"
        elif probes_bs < probes_is:
            winner = "Binary"
        else:
            winner = "Tie"

        print(f"{target:>16} | {probes_is:>10} | {probes_bs:>10} | {winner:>12}")

# --- Main Execution ---
if __name__ == "__main__":
    compare_probe_performance()
```
