```python
import random

def interpolation_search_floats(arr, target):
    """
    Interpolation Search adapted for floating-point numbers.
    Returns: (index_found, number_of_comparisons)
    """
    low = 0
    high = len(arr) - 1
    comparisons = 0

    while low <= high and arr[low] <= target <= arr[high]:
        comparisons += 1

        if low == high:
            if arr[low] == target:
                return low, comparisons
            return -1, comparisons

        # Guard against division by zero if elements are identical
        if arr[high] == arr[low]:
            if arr[low] == target:
                return low, comparisons
            return -1, comparisons

        # Interpolation formula handles the float subtraction,
        # then casts the final offset to an integer index.
        pos = low + int(((target - arr[low]) * (high - low)) /
                        (arr[high] - arr[low]))

        # Ensure pos stays within bounds (sometimes floating point math
        # can push it slightly out of bounds in extreme edge cases)
        pos = min(max(pos, low), high)

        if arr[pos] == target:
            return pos, comparisons
        elif arr[pos] < target:
            low = pos + 1
        else:
            high = pos - 1

    return -1, comparisons


def binary_search(arr, target):
    """
    Binary Search for comparison.
    Returns: (index_found, number_of_comparisons)
    """
    low = 0
    high = len(arr) - 1
    comparisons = 0

    while low <= high:
        comparisons += 1
        mid = (low + high) // 2

        if arr[mid] == target:
            return mid, comparisons
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1

    return -1, comparisons


def float_performance_analysis():
    """
    Analyzes the number of comparisons needed for Interpolation Search
    vs Binary Search on uniformly distributed floating-point numbers.
    """
    sizes = [10000, 50000, 100000]
    num_trials = 100 # Average over 100 searches for stable data

    print("--- Floating-Point Search Comparison Analysis ---")
    print("Range: 0.0 to 1000.0 (Uniformly Distributed)\n")
    print(f"{'Dataset Size':>12} | {'Avg IS Comparisons':>20} | {'Avg BS Comparisons':>20}")
    print("-" * 60)

    for size in sizes:
        # Generate uniformly distributed floating-point numbers
        arr = sorted([random.uniform(0.0, 1000.0) for _ in range(size)])

        total_is_comps = 0
        total_bs_comps = 0

        for _ in range(num_trials):
            # Pick a random valid target from the array
            target = arr[random.randint(0, size - 1)]

            _, comps_is = interpolation_search_floats(arr, target)
            _, comps_bs = binary_search(arr, target)

            total_is_comps += comps_is
            total_bs_comps += comps_bs

        # Calculate averages
        avg_is = total_is_comps / num_trials
        avg_bs = total_bs_comps / num_trials

        print(f"{size:>12} | {avg_is:>20.2f} | {avg_bs:>20.2f}")

# --- Main Execution ---
if __name__ == "__main__":
    # 1. Quick isolated test
    sample_arr = sorted([random.uniform(0.0, 1000.0) for _ in range(15)])
    sample_target = sample_arr[7] # Pick the middle element

    idx, comps = interpolation_search_floats(sample_arr, sample_target)

    print("--- Quick Function Test ---")
    print(f"Target Float : {sample_target:.4f}")
    print(f"Found at idx : {idx}")
    print(f"Comparisons  : {comps}\n")

    # 2. Run the full analysis
    float_performance_analysis()
```
